```python
covid_data_df = pd.read_csv('covid-countries-data.csv')
covid_data_df
```

total missing test

```python
covid_data_df['total_tests'].isna().sum()
```

merge two csv file countries_df and covid_data_df on location column

```python
mearged_df = countries_df.merge(covid_data_df, on='location')
mearged_df
```

add column tests_per_million, cases_per_million and deaths_per_million into combined_df.



```python
mearged_df['tests_per_million'] = mearged_df['total_tests'] * 1e6 / mearged_df['population']
mearged_df
```

```python
mearged_df['cases_per_million'] = mearged_df['total_cases'] * 1e6 / mearged_df['population']
mearged_df
```

```python
mearged_df['deaths_per_million'] = mearged_df['total_deaths'] * 1e6 / mearged_df['population']
mearged_df
```

10 countries with highest test per million

```python
highest_test_per_million = mearged_df.sort_values('tests_per_million', ascending=False).head(10)
highest_test_per_million
```

10 countries with  highest number of positive cases per million people.

```python
highest_positive_cases_per_million = mearged_df.sort_values('cases_per_million', ascending=False).head(10)
highest_positive_cases_per_million
```

10 countires that have highest number of deaths cases per million people

```python
highest_death_cases_per_million = mearged_df.sort_values('deaths_per_million', ascending=False).head(10)
highest_death_cases_per_million
```

number of countries that feature in both the lists of "highest number of tests per million" and "highest number of cases per million"

```python
both_df = pd.merge(highest_test_per_million, highest_positive_cases_per_million, on='location')
both_df
```

```python
len(both_df)
print (f'{len(both_df)} countries feature in both the lists of highest number of tests per million and highest number of cases per million')
```

number of countries that feature in both the lists "20 countries with lowest GDP per capita" and "20 countries with the lowest number of hospital beds per thousand population". Only consider countries with a population higher than 10 million while creating the list.

```python
#country filter with population more than 10 million
filter_country = mearged_df[mearged_df['population'] > 10000000]
filter_country
```

```python
#20 countries with lowesr gdp

lowest_gdp = filter_country.sort_values('gdp_per_capita').nsmallest(20, 'gdp_per_capita')
lowest_gdp
```

```python
#20 countries with lowest bed count
lowest_hospital_beds = filter_country.sort_values('hospital_beds_per_thousand').nsmallest(20, 'hospital_beds_per_thousand')
lowest_hospital_beds
```

```python
common = pd.merge(lowest_gdp, lowest_hospital_beds, on='location')
common
```

```python
number = len(common)
print(f'{number} countries feature in both the lists')
```
